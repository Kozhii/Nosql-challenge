5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Define the query condition to find documents where latitude and longitude are stored as strings
query = {
    "$or": [
        {"geocode.latitude": {"$type": "string"}},
        {"geocode.longitude": {"$type": "string"}}
    ]
}

# Define the update operation to convert "geocode.latitude" and "geocode.longitude" to decimal numbers
update_operation = [
    {
        "$set": {
            "geocode.latitude": {
                "$toDouble": "$geocode.latitude"
            },
            "geocode.longitude": {
                "$toDouble": "$geocode.longitude"
            }
        }
    }
]

# Use update_many to apply the update operation to matching documents
update_result = establishments.update_many(query, update_operation)

if update_result.modified_count > 0:
    print(f"Converted latitude and longitude in {update_result.modified_count} documents to decimal numbers.")
else:
    print("No documents were converted.")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue

# Define the query to match all documents
query = {}

# Define the update operation to change the data type of "RatingValue" to an integer
update_operation = [
    {
        "$set": {
            "RatingValue": {
                "$cond": {
                    "if": { "$eq": [ { "$type": "$RatingValue" }, "string" ] },
                    "then": { "$toInt": "$RatingValue" },
                    "else": "$RatingValue"
                }
            }
        }
    }
]

# Use update_many to apply the update operation to all documents
update_result = establishments.update_many(query, update_operation)

if update_result.modified_count > 0:
    print(f"Successfully changed the data type of 'RatingValue' to integer in {update_result.modified_count} documents.")
else:
    print("No documents were updated.")

```

```python
# Check that the coordinates and rating value are now numbers
# Find a single document from the collection
document = establishments.find_one()

if document:
    # Extract and check the data types and values of the latitude, longitude, and rating fields
    latitude = document.get("geocode", {}).get("latitude")
    longitude = document.get("geocode", {}).get("longitude")
    rating_value = document.get("RatingValue")

    latitude_type = type(latitude)
    longitude_type = type(longitude)
    rating_value_type = type(rating_value)

    print(f"Latitude Type: {latitude_type}, Latitude Value: {latitude}")
    print(f"Longitude Type: {longitude_type}, Longitude Value: {longitude}")
    print(f"Rating Value Type: {rating_value_type}, Rating Value: {rating_value}")
else:
    print("No documents found in the collection.")
```
